```python
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor

X = 2* np.random.rand(100,1)
y= 4+3 * X + np.random.randn(100,1)

lin_reg = LinearRegression()
lin_reg.fit(X,y)

print(f"normal eq intercept: {lin_reg.intercept_} , {lin_reg.coef_}")













sgd_reg = SGDRegressor(max_iter=1000, tol = 1e-3, eta0 = 0.1, random_state = 42)

sgd_reg.fit(X, y.ravel())

print(f"sdg intercept: {sgd_reg.intercept_} and coef: {sgd_reg.coef_}")
```

```python
from sklearn.linear_model._base import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
import numpy as np

X = 6 * np.random.rand(100, 1) - 3
Y = 0.5 * X**2 + X+2 + np.random.randn(100, 1)

poly = Pipeline(
    [
        ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
        ("lin_reg", LinearRegression())
    ]
)

poly.fit(X,y)
print(f"model coeffs: {poly.named_steps['lin_reg'].coef_}")
```
