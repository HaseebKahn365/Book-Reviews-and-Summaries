```python
import numpy as np

def sigmoid(x):
    """Implement sigmoid activation function."""
    return 1. / (1 + np.exp(-x))

def dense_layer(x, w, b):
    y = x@w + b
    return sigmoid(y)
# Simple demo
x = np.array([[1, 2], [3, 4]])
w = np.array([[0.5, 0.3], [0.2, 0.4]])
b = np.array([0.1, 0.2])

print("Input x:\n", x)
print("\nWeights w:\n", w)
print("\nBias b:\n", b)

result = dense_layer(x, w, b)
print("\nOutput:\n", result)
```
