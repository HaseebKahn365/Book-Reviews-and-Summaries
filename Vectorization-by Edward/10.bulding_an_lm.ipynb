```python
import pandas as pd
import numpy as np


sales_data = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104],
    'Price': [25.0, np.nan, 15.0, 40.0]
})

inventory_data = pd.DataFrame({
    'Product_ID': [102, 103, 104, 105],
    'Stock': [50, 30, 20, 10]
})


merged_df = pd.merge(sales_data, inventory_data, on='Product_ID')


mean_price = merged_df['Price'].mean()
merged_df['Price'] = merged_df['Price'].fillna(mean_price)


outliers = merged_df[merged_df['Price'] > 30]

print("Cleaned & Aligned Data:\n", merged_df)
print("\nOutliers Found:\n", outliers)
```
