```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("retail_data.csv")
df.head()
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.histplot(df["Annual_Income"], kde= True)
plt.title("Skewed Distribution of income")
```

```python
plt.subplot(2,2,2)
sns.boxplot(x=df["Spending_Score"])
plt.title("Plot for spending via boxplot")
plt.show()
```

```python
q1 = df["Annual_Income"].quantile(0.25)
q3 = df["Annual_Income"].quantile(0.75)

iqr = q3 - q1

up = q3+1.5 * iqr
lower = q1- 1.5* iqr


print("left outliers:")
print(df[df["Annual_Income"]<lower ]  )

print(f"Right outliers ({up})")
print(df[df["Annual_Income"]>up])

```
