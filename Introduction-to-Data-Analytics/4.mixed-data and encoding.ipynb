```python
import pandas as pd
import numpy as np

df = pd.read_csv('retail_data.csv')
df
```

```python
df['Mixed_ID'] = pd.to_numeric(df['Mixed_ID'], errors = 'coerce')
df['Mixed_ID'].sample(10)
```

```python
df['Mixed_ID'].fillna(0, inplace=True)
```

```python
df['cleaned_dates'] = pd.to_datetime(df['Transaction_Date'], errors = 'coerce')
```

```python
df.sample(19)
```

```python
df['year'] = df['cleaned_dates'].dt.year
```

```python

from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

ordenc =OrdinalEncoder()
df['City_Ordinal'] = ordenc.fit_transform(df[['City']].fillna('Unknown'))
df.head()

labenc = LabelEncoder()
df['City_Label'] = labenc.fit_transform(df[['City']].fillna("unknown"))


df_ohe = pd.get_dummies(df, columns= ['City'], prefix ="is" )
df_ohe
```
