```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('retail_data.csv')
df.head()
```

```python
df['Annual_Income'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data= df, x='City', palette= 'viridis')
total = len(df['City'].dropna())
for p in ax.patches:
    perc= f'{100 * p.get_height() / total:.1f} %'
    x = p.get_x() +p.get_width() / 2 - 0.15
    y = p.get_height() +0.5
    ax.annotate(perc, (x,y), fontsize= 10, fontweight= 'bold')

plt.title("frequrency and percentage")
plt.show()
```

# Sub-task 4: Detect Noisy Data (Visualizing Outliers as Noise)


```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Spending_Score'], color = 'salmon')
plt.title("noisy data box plot")
plt.show()

print("skewness:")
sns.histplot(df['Annual_Income'], kde=True, color= 'teal')
plt.title("Skewness visualized: ")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(data= df, x= 'Age',y='Spending_Score', scatter_kws = {'alpha': 0.5}, line_kws= {'color' : 'red'})
plt.title('Bivariate analysis: age vs spending')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Annual_Income', 'Spending_Score', 'City']])
plt.suptitle('Multivariate analysis, ',y= 1.02)
plt.show()
```
