```python
class Node:
    def __init__(self, value):
        self._value = value
        self._children = []
    def value(self):
        return self._value

    def __repr__(self):
        return 'Node({!r})'.format(self._value)

    def add_child(self, node):
        self._children.append(node)

    def __iter__(self):
        return iter(self._children)

def print_tree(node,depth = 5):
    print(f"{' ' * depth}Node value: {node.value()}")
    for c in node:
        print_tree(c, depth+5)
```

```python
root = Node(0)
child1 = Node(1)
child2 = Node(2)
child11 = Node(11)

root.add_child(child1)
child1.add_child(child11)
root.add_child(child2)

print_tree(root)
```

```python
def frange(start, stop, increment =1):
    x = start
    while x<stop:
        yield x
        x+=increment
```

```python
for n in frange(21,31, 0.5):
    print(n)
```

```python
# Depth first iterator:
class Node:
    def __init__(self, value):
        self._value = value
        self._children = []

    def __repr__(self):
        return 'Node({!r})'.format(self._value)

    def add_child(self, node):
        self._children.append(node)

    def __iter__(self):
        return iter(self._children)

    def depth_first(self):
        yield self
        for c in self:
            yield from c.depth_first()
```

```python
root = Node(0)
child1 = Node(1)
child2 = Node(2)
root.add_child(child1)
root.add_child(child2)
child1.add_child(Node(3))
child1.add_child(Node(4))
child2.add_child(Node(5))

for ch in root.depth_first():
    print(ch)
```

```python
from collections import deque
class linehistory:
    def __init__(self, lines, histlen=3):
        self.lines = lines
        self.history = deque(maxlen=histlen)
        
    def __iter__(self):
        for lineno, line in enumerate(self.lines, 1):
            self.history.append((lineno, line))
            yield line

    def clear(self):
        self.history.clear()

    def show_history(self):
        for l in self.history:
            print(l)
```

```python
with open('3.nums_dates_and_time.ipynb') as f:
    lines = linehistory(f)
    for line in lines:
        if 'python' in line:
            for lineno, hline in lines.history:
                print('{}:{}'.format(lineno, hline), end='')
    
    print("\n\n")
    lines.show_history()
```

```python
import os
import fnmatch
import gzip
import bz2
import re

def gen_find(filepat, top):
    '''
    Find all filenames in a directory tree that match a shell wildcard pattern
    '''
    for path, dirlist, filelist in os.walk(top):
        for name in fnmatch.filter(filelist, filepat):
            yield os.path.join(path, name)

def gen_opener(filenames):
    '''
    Open a sequence of filenames one at a time producing a file object.
    The file is closed immediately when proceeding to the next iteration.
    '''
    for filename in filenames:
        if filename.endswith('.gz'):
            f = gzip.open(filename, 'rt')
        elif filename.endswith('.bz2'):
            f = bz2.open(filename, 'rt')
        else:
            f = open(filename, 'rt')
        yield f
        f.close()

def gen_concatenate(iterators):
    '''
    Chain a sequence of iterators together into a single sequence.
    '''
    for it in iterators:
        yield from it

def gen_grep(pattern, lines):
    '''
    Look for a regex pattern in a sequence of lines
    '''
    pat = re.compile(pattern)
    for line in lines:
        if pat.search(line): yield line
```

```python
# --- STEP 1: Define the directory to search ---
current_dir = "." # Current directory

# --- STEP 2: Build the Pipeline ---
# 1. Find all .ipynb files
files = gen_find("*.ipynb", current_dir)

# 2. Open them
files_obj = gen_opener(files)

# 3. Concatenate all lines into one stream
lines = gen_concatenate(files_obj)

# 4. Grep for the pattern "python"
# (Using a case-insensitive search for the demo)
matches = gen_grep("(?i)python", lines)

# --- STEP 3: Execute the Pipeline ---
print("--- SEARCH RESULTS ---")
count = 0
for line in matches:
    print(line.strip()[:] + "...") # Truncating for readability
   

if count == 0:
    print("No matches found.")
```
