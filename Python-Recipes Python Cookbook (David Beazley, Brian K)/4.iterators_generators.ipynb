```python
class Node:
    def __init__(self, value):
        self._value = value
        self._children = []
    def value(self):
        return self._value

    def __repr__(self):
        return 'Node({!r})'.format(self._value)

    def add_child(self, node):
        self._children.append(node)

    def __iter__(self):
        return iter(self._children)

def print_tree(node,depth = 5):
    print(f"{' ' * depth}Node value: {node.value()}")
    for c in node:
        print_tree(c, depth+5)
```

```python
root = Node(0)
child1 = Node(1)
child2 = Node(2)
child11 = Node(11)

root.add_child(child1)
child1.add_child(child11)
root.add_child(child2)

print_tree(root)
```

```python
def frange(start, stop, increment =1):
    x = start
    while x<stop:
        yield x
        x+=increment
```

```python
for n in frange(21,31, 0.5):
    print(n)
```

```python
# Depth first iterator:
class Node:
    def __init__(self, value):
        self._value = value
        self._children = []

    def __repr__(self):
        return 'Node({!r})'.format(self._value)

    def add_child(self, node):
        self._children.append(node)

    def __iter__(self):
        return iter(self._children)

    def depth_first(self):
        yield self
        for c in self:
            yield from c.depth_first()
```

```python
root = Node(0)
child1 = Node(1)
child2 = Node(2)
root.add_child(child1)
root.add_child(child2)
child1.add_child(Node(3))
child1.add_child(Node(4))
child2.add_child(Node(5))

for ch in root.depth_first():
    print(ch)
```

```python
from collections import deque
class linehistory:
    def __init__(self, lines, histlen=3):
        self.lines = lines
        self.history = deque(maxlen=histlen)
        
    def __iter__(self):
        for lineno, line in enumerate(self.lines, 1):
            self.history.append((lineno, line))
            yield line

    def clear(self):
        self.history.clear()

    def show_history(self):
        for l in self.history:
            print(l)
```

```python
with open('3.nums_dates_and_time.ipynb') as f:
    lines = linehistory(f)
    for line in lines:
        if 'python' in line:
            for lineno, hline in lines.history:
                print('{}:{}'.format(lineno, hline), end='')
    
    print("\n\n")
    lines.show_history()
```
