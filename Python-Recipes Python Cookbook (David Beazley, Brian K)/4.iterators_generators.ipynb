```python
class Node:
    def __init__(self, value):
        self._value = value
        self._children = []

    def __repr__(self):
        return 'Node({!r})'.format(self._value)

    def add_child(self, node):
        self._children.append(node)

    def __iter__(self):
        return iter(self._children)
```

```python
root = Node(0)
child1 = Node(1)
child2 = Node(2)
child11 = Node(11)

root.add_child(1)
child1.add_child(child11)
root.add_child(child2)

for c in root:
    print(c)
```
