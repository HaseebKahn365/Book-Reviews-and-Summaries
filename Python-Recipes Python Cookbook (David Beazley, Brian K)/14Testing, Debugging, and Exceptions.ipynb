The "Stunt Double" (Mocking & Patching)

```python
from unittest import TestCase
from unittest.mock import MagicMock, patch

def delete_user(db_conn, user_id):
    querry = f"del from users where id = {user_id}"
    db_conn.execute(querry)

class TestDatabase(TestCase):
    def test_del(self):
        mock_db = MagicMock()
        delete_user(mock_db, 42)

        mock_db.execute.assert_called_with("del from users where id = 42")


t = TestDatabase()
t.test_del()
```
