The "Stunt Double" (Mocking & Patching)

```python
from unittest import TestCase
from unittest.mock import MagicMock, patch

def delete_user(db_conn, user_id):
    querry = f"del from users where id = {user_id}"
    db_conn.execute(querry)

class TestDatabase(TestCase):
    def test_del(self):
        mock_db = MagicMock()
        delete_user(mock_db, 42)

        mock_db.execute.assert_called_with("del from users where id = 42")


t = TestDatabase()
t.test_del()
```

```python
import unittest

# The "Silent Auditor" (Logging & Skipping)
import unittest
import sys
import os

class MyUtilT(unittest.TestCase):
    @unittest.skipIf(os.name != 'nt', "Only runs on windows")
    def test_win_reg(self):
        pass

    @unittest.expectedFailure
    def test_newf(self):
        self.assertEqual(1,0)

def run_n_log():
    suite = unittest.TestLoader().loadTestsFromModule(sys.modules[__name__])

    with open('test_rep.txt', 'w') as f:
        runner = unittest.TextTestRunner(stream=f, verbosity=2, descriptions=True)
        runner.run(suite)
run_n_log()
```
